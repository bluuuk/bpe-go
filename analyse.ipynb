```python
from pathlib import Path 
from collections import defaultdict,Counter
import base64
```

```python
with Path.cwd() / "cl100k_base.tiktoken" as f:
    data = f.read_text()
```

```python
c = Counter()
for line in data.splitlines():
    tok,rank = line.split(" ",1)
    tok = base64.b64decode(tok).decode(errors="replace")
    c[tok] = int(rank) 
```

```python
paths = list(
    word for word in c.keys() if not set(word).intersection("./=*()[]{}#-_%+\n~") and len(word) != word.count(" ")   
)
paths.sort(reverse=True,key=len)
paths
```
